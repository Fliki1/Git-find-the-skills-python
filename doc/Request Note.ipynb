### Download image

Se si vuole scaricare un immagine invece, si usa il content, che ritorna il contenuto della risposta in byte
Si utilizza l'url dell'immagine ovviamente

```python
i = requests.get('https://imgs.xkcd.com/comics/dorm_poster.png')
print(i.content)
```

Otteniamo il byte di quella immagine, se salviamo questi in un file nella macchina otteniamo l'immagine

```python
i = requests.get('https://imgs.xkcd.com/comics/dorm_poster.png')

with open('comic.png', 'wb') as f: # apriamo un file in writebyte
    f.write(i.content)
```

Proviamo a usare una chiamata url con dei parametri richiesti

per inserire i parametri si potrebbe scrivere direttamente nell'url tipo ?=page2 ecc..
ma requests mette a disposizione la possibilità di inserire questi parametri come dict python

```python
payload = {'page': 2, 'count': 25}
r = requests.get('https://httpbin.org/get', params=payload)
print(r.text)
```

## JSON risposta

Stiamo sempre ottenendo una JSON come risposta, anzichè quindi usare il .text come metodo usiamo il metodo json

```python
payload = {'username': 'corey', 'password': 'testing'}
r = requests.post('https://httpbin.org/post', data=payload)
print(r.json())
```

Ha creato un Python dict della JSON response, provo a salvarla in una variabile e discutiamone

```python
payload = {'username': 'corey', 'password': 'testing'}
r = requests.post('https://httpbin.org/post', data=payload)

r_dict = r.json()

print(r_dict['form'])
```

Come si può vedere possiamo usare e cercare es. accedere al contenuto del "form" value

## Auth

Per testare l'auth ci necessita di invocare un altro url della forma: https://httpbin.org/basic-auth/{user}/{password}

i parametri passati in quel url saranno presi come buoni, cosi che, se rinvocati con le stesse credenziali, ritornerà un esito true di access, false altrimenti

```python
r = requests.get('https://httpbin.org/basic-auth/corey/testing', auth=('corey','testing'))

print(r.text)
```

Otteniamo la risposta di authenticazione avvenuta, perché stiamo accendendo con auth con credenziali che reputa vere il sito di testing httpbin.org

```python
r = requests.get('https://httpbin.org/basic-auth/corey/testing', auth=('corley','testing'))

print(r)
```

Abbiamo una risposta di status di errore 400, che non è autorizzato ad avere un response code, per questo non abbiamo r.text, ma r direttamente

```python
r = requests.get('https://httpbin.org/basic-auth/corey/testing', auth=('corey','testing'))

print(r)
```

## Timer

Per monitorare un sito per esempio si cerca di avere una risposta a una richiesta che si sta aspettando da tempo magari, avere un timeout da settare per essere sicuri che il server potrà rispondere o meno sarebbe ottimo. Un timer di attesa minimo per dire che ci garantisca che non stiamo aspettando a vuoto chissà per quanto.

Si aggiunge un semplice parametro

```python
r = requests.get('https://httpbin.org/delay/1', timeout=3)

print(r)
```

Per fare simulare un delay dal sito, per testare il timeout https://httpbin.org/delay/{secondi}

Sopra abbiamo un delay di 1sec e un timeout di 3sec

```python
r = requests.get('https://httpbin.org/delay/6', timeout=3)

print(r)
```
